# Frekuensi Take Task

Frekuensi take task untuk seluruh employee

```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd

# Pastikan path file sudah benar
file_path = '/content/drive/My Drive/Database Platon Tahun 2024.csv'

# Baca data dari file CSV
df = pd.read_csv(file_path)

# Asumsikan kolom 'datetime_column' berisi date dan time jadi satu
datetime_column = 'takenDate'  # Ganti dengan nama kolom yang benar

# 1. Ubah kolom datetime menjadi tipe data datetime, handle error dengan coerce
df[datetime_column] = pd.to_datetime(df[datetime_column], errors='coerce')

# 2. Drop baris dengan nilai NaT (Not a Time) di kolom 'takenDate'
df.dropna(subset=[datetime_column], inplace=True)

# 3. Ekstrak hanya jam
df['hour'] = df[datetime_column].dt.hour  # Jam (0-23)

# 4. Hitung frekuensi hanya berdasarkan jam
frequency_data = df.groupby('hour').size().reset_index(name='frequency')

# Tampilkan data frekuensi
print(frequency_data)

# Atau, untuk tampilan yang lebih rapi di notebook:
display(frequency_data)
```

```python
from matplotlib import pyplot as plt
frequency_data.plot(kind='scatter', x='hour', y='frequency', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
frequency_data['frequency'].plot(kind='line', figsize=(8, 4), title='frequency')
plt.gca().spines[['top', 'right']].set_visible(False)
```
