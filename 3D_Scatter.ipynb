```python
!pip install plotly
import plotly.express as px
import pandas as pd

# Assuming 'df' is your dataframe with 'endDate' and 'tanggal_takenDate' columns

# 1. Menghitung Frekuensi
endDate_freq = df['endDate'].value_counts().reset_index()
endDate_freq.columns = ['endDate', 'endDate_freq']

takenDate_freq = df['tanggal_takenDate'].value_counts().reset_index()
takenDate_freq.columns = ['tanggal_takenDate', 'takenDate_freq']

# 2. Mempersiapkan Data untuk Plot
plot_df = pd.merge(endDate_freq, takenDate_freq,
                   left_on='endDate', right_on='tanggal_takenDate', how='outer')
plot_df.fillna(0, inplace=True)  # Mengisi nilai NaN dengan 0

# Mengubah tipe data kolom tanggal ke datetime
plot_df['endDate'] = pd.to_datetime(plot_df['endDate'])
plot_df['tanggal_takenDate'] = pd.to_datetime(plot_df['tanggal_takenDate'])

# Mengambil komponen tanggal (hari, bulan, tahun)
plot_df['day'] = plot_df['endDate'].dt.day.fillna(plot_df['tanggal_takenDate'].dt.day).astype(int)
plot_df['month'] = plot_df['endDate'].dt.month.fillna(plot_df['tanggal_takenDate'].dt.month).astype(int)
plot_df['year'] = plot_df['endDate'].dt.year.fillna(plot_df['tanggal_takenDate'].dt.year).astype(int)

# 3. Membuat Plot 3D
fig = px.scatter_3d(plot_df, x='day', y='month', z='year',
                    color='endDate_freq', size='takenDate_freq',
                    hover_data=['endDate', 'tanggal_takenDate'],
                    labels={'endDate_freq': 'Frequency endDate',
                            'takenDate_freq': 'Frequency takenDate'})

fig.update_layout(scene=dict(xaxis_title='Day',
                             yaxis_title='Month',
                             zaxis_title='Year'),
                  title='3D Scatter Plot of endDate vs tanggal_takenDate Frequency')

fig.show()
```

```python
!pip install plotly
import plotly.express as px
import pandas as pd

# Assuming 'df' is your dataframe with 'endDate' and 'tanggal_takenDate' columns

# 1. Filter data untuk tahun 2024-2025
filtered_df = df[
    (df['endDate'].dt.year >= 2024) & (df['endDate'].dt.year <= 2025)
]

# 2. Menghitung Frekuensi
endDate_freq = filtered_df['endDate'].value_counts().reset_index()
endDate_freq.columns = ['endDate', 'endDate_freq']

takenDate_freq = filtered_df['tanggal_takenDate'].value_counts().reset_index()
takenDate_freq.columns = ['tanggal_takenDate', 'takenDate_freq']

# 3. Mempersiapkan Data untuk Plot
plot_df = pd.merge(
    endDate_freq, takenDate_freq, left_on='endDate', right_on='tanggal_takenDate', how='outer'
)
plot_df.fillna(0, inplace=True)  # Mengisi nilai NaN dengan 0

# Ensure 'endDate' column is of datetime type before extracting month
# If it's not, try converting it:
plot_df['endDate'] = pd.to_datetime(plot_df['endDate'], errors='coerce')

# 4. Membuat Plot 3D (dengan bulan sebagai sumbu x, frekuensi sebagai sumbu y dan z)
fig = px.scatter_3d(
    plot_df,
    x=plot_df['endDate'].dt.month,
    y='endDate_freq',
    z='takenDate_freq',
    hover_data=['endDate', 'tanggal_takenDate'],
    labels={'x': 'Month', 'y': 'Frequency endDate', 'z': 'Frequency takenDate'},
)

fig.update_layout(
    scene=dict(
        xaxis_title='Month', yaxis_title='Frequency endDate', zaxis_title='Frequency takenDate'
    ),
    title='3D Scatter Plot of endDate vs tanggal_takenDate Frequency (2024-2025)',
)

fig.show()
```

```python
!pip install plotly
import plotly.express as px
import pandas as pd

# Assuming 'df' is your dataframe with 'endDate' and 'tanggal_takenDate' columns

# 1. Filter data untuk tahun 2024-2025
filtered_df = df[(df['endDate'].dt.year >= 2024) & (df['endDate'].dt.year <= 2025)]

# 2. Menghitung Frekuensi
endDate_freq = filtered_df['endDate'].value_counts().reset_index()
endDate_freq.columns = ['Date', 'Frequency']
endDate_freq['Date_Type'] = 'endDate'

takenDate_freq = filtered_df['tanggal_takenDate'].value_counts().reset_index()
takenDate_freq.columns = ['Date', 'Frequency']
takenDate_freq['Date_Type'] = 'tanggal_takenDate'

# 3. Menggabungkan Data untuk Plot
plot_df = pd.concat([endDate_freq, takenDate_freq])

# 4. Mengubah tipe data kolom tanggal ke datetime
plot_df['Date'] = pd.to_datetime(plot_df['Date'], errors='coerce')

# 5. Membuat Plot 3D (dengan bulan sebagai sumbu x, frekuensi sebagai sumbu y dan z)
fig = px.scatter_3d(
    plot_df,
    x=plot_df['Date'].dt.month,
    y='Frequency',
    z=plot_df['Date_Type'],  # Menggunakan Date_Type sebagai sumbu z untuk pemisahan
    color='Date_Type',  # Memberi warna berdasarkan Date_Type
    hover_data=['Date', 'Frequency'],
    labels={'x': 'Month', 'y': 'Frequency', 'z': 'Date Type'},
)

fig.update_layout(
    scene=dict(
        xaxis_title='Month',
        yaxis_title='Frequency',
        zaxis_title='Date Type',
    ),
    title='3D Scatter Plot of endDate vs tanggal_takenDate Frequency (2024-2025)',
)

fig.show()
```
